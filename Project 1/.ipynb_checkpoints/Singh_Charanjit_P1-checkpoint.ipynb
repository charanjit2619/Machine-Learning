### No xrange

The xrange from Python 2 is now merged into "range" for Python 3 and there is no xrange in Python 3. In Python 3, range(3) does not create a list of 3 elements as it would in Python 2, rather just creates a more memory efficient iterator.

Hence,
xrange in Python 3: Does not exist
range in Python 3: Has very similar behavior to Python 2's xrange

````python
for i in range(3):
    print(i)
````

# Project Description

## Data File

The first line of `IrisData.txt` contains two integers. The first represents how many lines of data are in the file. The second how many features are associated with each line. Each line after that contains four floating point values representing the sepal length, sepal width, petal length, and petal width of a type of iris flower, followed by the name of the iris type: either setosa, versicolor, or virginica. Values are tab separated.

## Python Program

Write a Python program that will print out a **two-dimensional plot** of any two features for all three varieties of iris flowers using the features of `matplotlib.pyplot`. Your program should begin by asking for the name of the input file. For example:
```python
Enter the name of your data file: Irisdata.txt
```

````python
##############################################################################
#         TODO: Write the code for reading data from file                    #
##############################################################################
import matplotlib.pyplot as plt                        
import numpy as np  

file_name = input("Enter the name of data file: ")        # Taking file name as input from user
file_data = open(file_name,"r" )                          # Opening file in Accessmode = read only

feature_dict = {
    0 : "Sepal Length",
    1 : "Sepal Width",
    2 : "Petal Length",
    3 : "Petal Width"
}

first_row = file_data.readline()                          # First row in data gives the number of lines and features
first_row = first_row.split()                             # Split function helps in making a list by removing whitespaces
rows = int(first_row[0])
features = int(first_row[1])

data = np.zeros([rows,5], dtype = object)                 # Defining a 'data' array containing Iris data 

for k in range(rows):
    row_string = file_data.readline()
    row_string = row_string.strip()                       # Strip function helps in removing leading and trailing whitespaces
    row_list = row_string.split("\t")
        
    for j in range(features+1):
        if j != 4:                                        # Adding first 4 rows of numbers as int in 'data' array
            data[k, j] = float(row_list[j])
        else:
            data[k,j] = row_list[j]                       # Adding the last row of type as str in 'data' array
    

file_data.close()                                         # Closing the file
  
##############################################################################
#                              END OF YOUR CODE                              #
##############################################################################
````

Then the program should prompt the user for two features to plot. For example, it could say:
```
You can do a plot of any two features of the Iris Data set The feature codes are:
0 = sepal length
1 = sepal width
2 = petal length
3 = petal width
```
```python
Enter feature code For the horizontal axis: 0
Enter feature code For the vertical axis: 1
```

````python
##############################################################################
#   TODO: Write the function for plotting using two features of the data     #
##############################################################################
def plot_feature_map():
    try:                                                  # Using try and except to account for errors in case of string input
        print("The features of Iris flower along with their codes are shown below:\nCode\tFeature")
        print("0\tSepal Length\n1\tSepal Width\n2\tPetal Length\n3\tPetal Width.\nPlease select two features to plot.")
        while True:                                       # This loop runs till both input values are valid 
            h_code = int(input("Enter feature code for horizontal axis: "))
            v_code = int(input("Enter feature code for vertical axis: "))
            if h_code < 4 and v_code < 4 and h_code >= 0 and v_code >= 0 :
                break
            else:
                print("Please select both values from --> 0,1,2,3")

        for i in range(rows):                             # Plotting the graph
            if data[i,4] == 'setosa':
                set_plot = plt.scatter(data[i, h_code], data[i, v_code], color = "red", marker = "o", label = "setosa")
            elif data[i,4] == "versicolor":
                versi_plot = plt.scatter(data[i, h_code], data[i, v_code], color = "yellow", marker = "v", label = "versicolor")
            elif data[i,4] == "virginica":
                vir_plot = plt.scatter(data[i, h_code], data[i, v_code], color = "blue", marker = "x", label = "virginica")

        plt.xlabel(f'{feature_dict[h_code]}')
        plt.ylabel(f'{feature_dict[v_code]}')
        plt.title("Iris Flower Feature Plot")
        plt.legend([set_plot, versi_plot, vir_plot],["setosa", "versicolor", "virginica"])
        plt.savefig('featureanalysis.png', bbox_inches='tight')
        plt.show() 
    except:
        print("Please check your input value. Allowed Inputs --> 0,1,2,3")
        
plot_feature_map()                                   # checking if function works properly
##############################################################################
#                              END OF YOUR CODE                              #
##############################################################################
````
